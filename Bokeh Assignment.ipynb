Q 2. glyphs are visual representations of your data points on a plot. They can be used to display various shapes, such as circles, squares, lines, bars, and more. Glyphs allow you to add interactivity and customization to your plots.

```python

import bokeh. plotting
bokeh.io.output_notebook()

from bokeh.plotting import figure, output_file, show

x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
output_file("bokeh_plot.html")

p = figure(title="My Bokeh Plot")

p.scatter(x,y,fill_color = 'red' , legend_label = 'red points',size = 20)
show(p)
```

Q 3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

```python
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
output_file("bokeh_plot.html")

p = figure(title="My Bokeh Plot")

p.scatter(x,y,fill_color = 'green' , legend_label = 'green points',size = 20)
show(p)
```

 Q 4. A Bokeh server is a Python server application that allows you to create and deploy interactive Bokeh plots and applications. It enables you to build web-based data visualization dashboards and applications that can be accessed and updated in real time by multiple users.

Using the Bokeh server, you can create interactive plots with widgets and tools that respond to user interactions such as mouse clicks, selections, or changes in widget values. When a user interacts with the plot or widget, the Bokeh server processes the event and updates the plot accordingly, providing a dynamic and responsive visualization experience.

To use the Bokeh server to create interactive plots that can be updated in real time, you need to follow these steps:

Import the necessary modules: Import the required modules from Bokeh and other relevant libraries, such as bokeh.plotting, bokeh.models, and bokeh.server.

Prepare the data and create the initial plot: Prepare your data and create the initial Bokeh plot using the usual Bokeh plotting functions. Set up any necessary widgets or tools that you want to include in your interactive plot.

Define the update callback function: Define a Python function that will be called whenever an event occurs that triggers a plot update. This function will contain the code to update the plot based on the user's interaction. The function should take arguments that represent the necessary objects for updating the plot, such as the plot object itself, data sources, or widget values.

Create the Bokeh application: Create a Bokeh application by wrapping your plot and the update callback function. This is done using the bokeh.server.application module. The application object defines how the Bokeh server should handle and process events.

Start the Bokeh server: Start the Bokeh server by running the bokeh serve command in your terminal or command prompt. Specify the Python script containing your Bokeh application as an argument to the command.
